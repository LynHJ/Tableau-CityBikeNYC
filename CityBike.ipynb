```python
import pandas as pd
```

### Create Bike stations list to correct id,lat,lng

```python
# Creat a empty dataframe to load Bike start Station details
BikeStartStation_df = pd.DataFrame(columns=['start_station_name','start_station_id','start_lat','start_lng'])

print(BikeStartStation_df)

# Creat a empty dataframe to load Bike end Station details
BikeEndStation_df = pd.DataFrame(columns=['end_station_name','end_station_id','end_lat','end_lng'])

print(BikeEndStation_df)
```

### 202209-citibike-tripdata.csv has the longest station list

```python
df09 = pd.read_csv('Resources/202209-citibike-tripdata.csv')

startstationlist=[i for i in df09['start_station_name'].unique()]
```

```python
df09_droped = df09.dropna(axis=0)

endstationlist=[j for j in df09_droped['end_station_name'].unique()]

BikeEndStation_df['end_station_name']=endstationlist
```

```python
BikeStartStation_df['start_station_name']=startstationlist

BikeEndStation_df['end_station_name']=endstationlist
```

```python
for i in range(0,len(BikeStartStation_df)):
    BikeStartStation_df.iloc[i,1]=df09.loc[df09['start_station_name']==BikeStartStation_df['start_station_name'][i]]['start_station_id'].value_counts().index[0]
    BikeStartStation_df.iloc[i,2]=df09.loc[df09['start_station_name']==BikeStartStation_df['start_station_name'][i]]['start_lat'].value_counts().index[0]
    BikeStartStation_df.iloc[i,3]=df09.loc[df09['start_station_name']==BikeStartStation_df['start_station_name'][i]]['start_lng'].value_counts().index[0]

BikeStartStation_df
```

```python
for i in range(0,len(BikeEndStation_df)):
    BikeEndStation_df.iloc[i,1]=df09.loc[df09['end_station_name']==BikeEndStation_df['end_station_name'][i]]['end_station_id'].value_counts().index[0]
    BikeEndStation_df.iloc[i,2]=df09.loc[df09['end_station_name']==BikeEndStation_df['end_station_name'][i]]['end_lat'].value_counts().index[0]
    BikeEndStation_df.iloc[i,3]=df09.loc[df09['end_station_name']==BikeEndStation_df['end_station_name'][i]]['end_lng'].value_counts().index[0]

BikeEndStation_df
```

## Combine Multiple CSV

```python

df = pd.concat(
    map(pd.read_csv, ['Resources/202206-citibike-tripdata.csv',
                      'Resources/202207-citibike-tripdata.csv',
                      'Resources/202208-citibike-tripdata.csv',
                      'Resources/202209-citibike-tripdata.csv']), ignore_index=True)
```

```python
df.head()
```

## Drop Duplicated Data

```python
cleaned_df =df. drop_duplicates()

cleaned_df
```

```python
cleaned_df.dtypes
```

## Adjust Corresponed Lat, Lng For Each Bike Station

```python
for i in range(0,len(BikeStartStation_df)):
    cleaned_df.loc[cleaned_df['start_station_name']==BikeStartStation_df.iloc[i,0],'start_station_id']=BikeStartStation_df.iloc[i,1]
    cleaned_df.loc[cleaned_df['start_station_name']==BikeStartStation_df.iloc[i,0],'start_lat']=BikeStartStation_df.iloc[i,2]
    cleaned_df.loc[cleaned_df['start_station_name']==BikeStartStation_df.iloc[i,0],'start_lng']=BikeStartStation_df.iloc[i,3]
```

```python
for i in range(0,len(BikeEndStation_df)):
    cleaned_df.loc[cleaned_df['end_station_name']==BikeEndStation_df.iloc[i,0],'end_station_id']=BikeEndStation_df.iloc[i,1]
    cleaned_df.loc[cleaned_df['end_station_name']==BikeEndStation_df.iloc[i,0],'end_lat']=BikeEndStation_df.iloc[i,2]
    cleaned_df.loc[cleaned_df['end_station_name']==BikeEndStation_df.iloc[i,0],'end_lng']=BikeEndStation_df.iloc[i,3]
```

## Add ['Trip Time']

```python
# Check Nan value
cleaned_df.loc[cleaned_df['ended_at'].isnull()]
```

```python
cleaned_df['Trip Time']=round((pd.to_datetime(cleaned_df['ended_at'])-pd.to_datetime(cleaned_df['started_at'])).dt.total_seconds()/60,2)
```

## Add ['Status']


```python
cleaned_df['Status']=""
```

```python
# 2~30 mins is vaild using time for casual and 2~45 mins for member
cleaned_df.loc[((cleaned_df['Trip Time']>=2) & (cleaned_df['Trip Time']<=30) & (cleaned_df['member_casual']=='casual')),['Status']]='Valid'
cleaned_df.loc[((cleaned_df['Trip Time']>=2) & (cleaned_df['Trip Time']<=45) & (cleaned_df['member_casual']=='member')),['Status']]='Valid'
# 30~45 mins is vaild using time for member
cleaned_df.loc[((cleaned_df['Trip Time']>30) & (cleaned_df['Trip Time']<1440) & (cleaned_df['member_casual']=='casual'))|\
            ((cleaned_df['Trip Time']>45) & (cleaned_df['Trip Time']<1440) & (cleaned_df['member_casual']=='member')),['Status']]='Over Time'
# over 1day usage, we have to investigate
cleaned_df.loc[(cleaned_df['Trip Time']>=1440),['Status']]='Investigate'
# Trip time <0 shows machine loads wrong time
cleaned_df.loc[(cleaned_df['Trip Time']<0),['Status']]='Timeout Error'
# Trip time between 0~2 mins indicates riderships are not trigered
cleaned_df.loc[((cleaned_df['Trip Time']>=0) & (cleaned_df['Trip Time']<2)),['Status']]='Unsuccessful Renting'
```

## Add ['Error']

```python
cleaned_df['Error Type']=""
```

```python


cleaned_df.loc[(cleaned_df['end_lat'].isnull()) & (cleaned_df['end_station_name'].isnull()),['Error Type']]='Data Loading Failed'
cleaned_df.loc[(cleaned_df['end_lat'].notnull()) & (cleaned_df['end_station_name'].isnull()),['Error Type']]='GPS Error'
cleaned_df.loc[(cleaned_df['end_lat'].notnull()) & (cleaned_df['end_station_name'].notnull()),['Error Type']]='Completed'
```
